```python
#data cleaning

#Get the date month into numbers
month_mapping = {
    'jan': 1,
    'feb': 2,
    'mar': 3,
    'apr': 4,
    'may': 5,
    'jun': 6,
    'jul': 7,
    'aug': 8,
    'sep': 9,
    'oct': 10,
    'nov': 11,
    'dec': 12
}

df.month = df.month.map(month_mapping)

#Get the date value into format of '01-12', '12-03'
df.day = df.day.astype(str).str.zfill(2)
df.month = df.month.astype(str).str.zfill(2)

df['date'] = df.day + '-' + df.month

df['date_month'] = pd.to_datetime('2024-' + df['date'], format='%Y-%d-%m', errors='coerce')\
                       .dt.strftime('%d-%m')

df.drop(['date', 'day', 'month'], axis=1, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Identifikasi kolom kategorikal (kecuali 'date_month')
cat_cols = [col for col in df.columns if df[col].dtype == 'object' and col != 'date_month']

# Tentukan jumlah kolom kategorikal
num_cols = len(cat_cols)

# Buat subplots hanya dengan jumlah yang diperlukan
fig, axs = plt.subplots((num_cols // 3) + (num_cols % 3 > 0), 3, figsize=(18, (num_cols // 3 + 1) * 4))
axs = axs.flatten()

# Loop melalui setiap kolom kategorikal dan buat bar plot horizontal
for i, col in enumerate(cat_cols):
    # Ambil nilai hitung untuk kolom
    dv = df[col].value_counts().reset_index()
    dv.columns = [col, 'count']  # Ganti nama kolom untuk kejelasan

    # Tentukan palet warna dengan jumlah warna yang sesuai
    unique_values = dv[col].nunique()
    palette = sns.color_palette("Set2", n_colors=unique_values)

    # Buat bar plot horizontal dengan palet warna dan parameter hue
    sns.barplot(data=dv, y=col, x='count', ax=axs[i], palette=palette, hue=col)
    axs[i].set_title(f'{col.title()} - In Data Distribution', fontsize=16)
    axs[i].set_xlabel('Count', fontsize=14)
    axs[i].set_ylabel(col.title(), fontsize=14)
    axs[i].tick_params(axis='x', labelsize=12)
    axs[i].tick_params(axis='y', labelsize=12)

# Hapus subplot yang tidak digunakan
for j in range(num_cols, len(axs)):
    fig.delaxes(axs[j])

# Sesuaikan tata letak untuk mencegah tumpang tindih
plt.tight_layout()
plt.show()
```
